2.What happens if stopwords are not removed?
Stop Words are the words like "is","the","and" etc. If we don't remove them, they can make data noisy because they don't add much meaning to the sentence.
This can slow down processing and also affect the performance of models, since important words get mixed with too many unimportant ones.

3.Provide two real-world scenarios where removing stopwords can be harmful.
Even though stopwords are often removed, there are cases where they are actually important.
for example:
1.Sentiment Analysis:
In a sentence like "I am not happy", the word "not" completely changes the meaning. If we remove it, the sentence becomes "happy", which gives the wrong sentiment.
2.Chatbots:
In a question like "What is your name?", removing stopwords leaves only "name", which loses the intent of the question.

4.Explain the difference between stemming and lemmatization.
Both stemming and lemmatization are used to reduce words to their base form, but they work differently.
Stemming simply cuts off parts of a word without worrying if the result is a real word. For example, "running" might become "runn", which isn’t actually correct.
Lemmatization, on the other hand, uses proper language rules and dictionaries. So "running" becomes "run", which is the correct base word.
In short, stemming is faster but less accurate, while lemmatization is slower but gives better results.

TASK-2

Build Advanced Preprocessing Function


```python
import re
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    #Lowercase
    text=text.lower()

    #Remove URLs & emails
    text=re.sub(r"http\S+|www\S+|https\S+", "",text)
    text=re.sub(r"\S+@\S+", "",text)

    #Remove numbers
    text=re.sub(r"\d+", "",text)

    #Handle repeated characters(soooo->soo)
    text=re.sub(r'(.)\1{2,}',r'\1\1',text)

    #Remove special characters/emojis
    text=re.sub(r"[^a-z\s]", "",text)

    #Remove extra spaces
    text=re.sub(r"\s+", " ",text).strip()
    tokens=text.split()

    #Remove short words (≤2) except "no", "not"
    cleaned_tokens=[
        word for word in tokens
        if len(word) > 2 or word in ["no","not"]
    ]

    cleaned_sentence=" ".join(cleaned_tokens)
    return cleaned_tokens,cleaned_sentence
```

TASK-3

Stress TEsting

```python
test_sentences=[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
for sentence in test_sentences:
    tokens,cleaned=preprocess_text(sentence)
    print("\nOriginal:",sentence)
    print("Tokens:",tokens)
    print("Cleaned:",cleaned)
```

TASK-4

Token Analytics


```python
def analyze_tokens(tokens):
    if not tokens:
        return 0,0,0
#calculating total,avg_length,unique tokens
    total=len(tokens)
    unique=len(set(tokens))
    avg_len=sum(len(t) for t in tokens)/total
    return total,unique,round(avg_len,2)
#iterating and calculating
for sentence in test_sentences:
    tokens,_ =preprocess_text(sentence)
    total,unique,avg=analyze_tokens(tokens)
#printing
    print("\nSentence:",sentence)
    print("Total Tokens:",total)
    print("Unique Tokens:",unique)
    print("Avg Length:",avg)
```

TASK-5

Frequency Analysis


```python
from collections import Counter
#initialzing empty list to store
all_tokens=[]
#iterating and checking
for sentence in test_sentences:
    tokens,_=preprocess_text(sentence)
    all_tokens.extend(tokens)
counter=Counter(all_tokens)
#printing top 10 and least 5 words
print("Top 10 words:",counter.most_common(10))
print("Least 5 words:",counter.most_common()[-5:])
```

TASk-6

Build Full Pipeline


```python
def full_pipeline(text_list):
  #empty list to store
    all_tokens=[]
    clean_sentences=[]
#iterating and appending
    for text in text_list:
        tokens,cleaned=preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
#printing
    print("\nAll Tokens:")
    print(all_tokens)
    print("\nCleaned Sentences:")
    for sentence in clean_sentences:
        print(sentence)
```
